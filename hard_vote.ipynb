# Directory 기준으로 파일 읽고 앙상블

```python
import pandas as pd

def create_answer(answer_dir):
    answer_csv = pd.read_csv(answer_dir)
    answers = []

    for idx in range(600):
        try:
            x = answer_csv.answers.iloc[idx]
            answers.append(x.split(','))
        except:
            answers.append('None')
    return answers

def predict_answer(answer, predict):
    
    predictions = list(predict.values())
    wrong_pair = []
    correct_pair = []
    for answer, predict in zip(answers, predictions):
        count = 0
        for ans in answer:
            if ans == predict:
                correct_pair.append((ans, predict))
                count += 1
        if count == 0:
            wrong_pair.append((answer, predict))
        
            
    # print(f'EM score {(600 - len(wrong_pair))/600}, 틀린 갯수: {len(wrong_pair)}')
    return (600 - len(wrong_pair))/600, predict
```

```python
import os
import json
from collections import Counter
from itertools import combinations,combinations_with_replacement
from tqdm import tqdm
def ensemble_with_hardvoting(answers, file_directory, output_directory,topk):
    """
    hard voting with prediction files
    Arguments:
        file_directory(str):
            str 로 이루어진 prediction 파일들이 있는 폴더 경로
        output_filename(str):
            str 로 이루어진 output 파일 경로+이x름
    Return:
        None
    """
    file_list = os.listdir(file_directory)
    json_dataset_list = {}


    for file in file_list:
        with open(os.path.join(file_directory, file)) as json_data:
            json_dataset_list[file] = json.load(json_data)

    total_score = []
    total_data = []
    total_output_name = []
    count = 1
    for k in tqdm(range(6,len(json_dataset_list)+1)):
        for ensemble_keys in combinations(json_dataset_list, k):
            for i,f_key in enumerate(ensemble_keys):
                ensembled_data = dict()
                for index in range(600):
                    temp_list = []
                    temp_list.append(list(json_dataset_list[f_key].values())[index])
                    for key in ensemble_keys[:i] + ensemble_keys[i+1:]:
                        temp_list.append(list(json_dataset_list[key].values())[index])
                    temp_counter = Counter(temp_list)
                    # print(temp_counter,temp_counter.most_common()[0][0])
                    ensembled_data[list(json_dataset_list[key].keys())[index]] = temp_counter.most_common()[0][0]

                score,data = predict_answer(answers,ensembled_data)
                total_score.append(score)
                total_data.append(data)
                total_output_name.append(os.path.join(output_directory,f'{score:.4f}|{count:10d}'+'.json'))
                count += 1

            
    final_index = sorted(range(len(total_score)), key=lambda k: -total_score[k])
    for i in final_index[:topk]:
        with open(total_output_name[i], 'w', encoding='UTF-8') as fp:
            json.dump(total_data[i], fp, indent=4, ensure_ascii=False)
```
